```python
import folium

m = folium.Map(
    location=(60.2, 24.8),
    zoom_start=10,
    control_scale=True
)

m
```

```python

WFS_BASE_URL = (
    "https://kartta.hel.fi/ws/geoserver/avoindata/wfs"
    "?service=wfs"
    "&version=2.0.0"
    "&request=GetFeature"
    "&srsName=EPSG:3879"
    "&typeName={layer:s}"
)

metro = (
    geopandas.read_file(
        WFS_BASE_URL.format(layer="avoindata:Seutukartta_liikenne_metro_rata")
    )
    .set_crs("EPSG:3879")
)
roads = (
    geopandas.read_file(
        WFS_BASE_URL.format(layer="avoindata:Seutukartta_liikenne_paatiet")
    )
    .set_crs("EPSG:3879")
)
```

```python
folium.GeoJson(roads, style_function=lambda x: {"color": "black", "opacity": "0.5"}).add_to(m)
folium.GeoJson(metro,style_function=lambda x: {"color": "red", "opacity": "0.7"}).add_to(m)
```
